```python
file_path = os.getcwd() + "\\vars_check_hyper_included\\levy_var_check_hyper_included_v1"
grids = []
acq_grids = []
hypers = []

for i in range(5,34):
    grids.append(np.load(file_path + "\\grid-%d.npy"%i))
    acq_grids.append(np.load(file_path + "\\acq_grid-%d.npy"%i))
    hypers.append(np.load(file_path+"\\hyperparameters-%d.npy"%i, allow_pickle=True))
```

```python
file_path = os.getcwd() + "\\vars_check_hyper_included\\levy_var_check_hyper_included_v5"
grids_v5 = []
acq_grids_v5 = []
hypers_v5 = []

for i in range(5,34):
    grids_v5.append(np.load(file_path + "\\grid-%d.npy"%i))
    acq_grids_v5.append(np.load(file_path + "\\acq_grid-%d.npy"%i))
    hypers_v5.append(np.load(file_path+"\\hyperparameters-%d.npy"%i, allow_pickle=True))
```

```python
file_path = os.getcwd() + "\\vars_check_hyper_included\\levy_var_check_hyper_included_v5_amp_v2"
grids_v5_v2 = []
acq_grids_v5_v2 = []
hypers_v5_v2 = []

for i in range(5,29):
    grids_v5_v2.append(np.load(file_path + "\\grid-%d.npy"%i))
    acq_grids_v5_v2.append(np.load(file_path + "\\acq_grid-%d.npy"%i))
    hypers_v5_v2.append(np.load(file_path+"\\hyperparameters-%d.npy"%i, allow_pickle=True))
```

```python
amp2s = []
lss = []

for item in hypers:
    amp2_list = [dict_['amp2'] for dict_ in item]
    ls_list = [dict_['ls'] for dict_ in item]
    amp2s.append(amp2_list)
    lss.append(ls_list)
```

```python
amp2s_v2 = []
lss_v2 = []

for item in hypers_v2:
    amp2_list = [dict_['amp2'] for dict_ in item]
    ls_list = [dict_['ls'] for dict_ in item]
    amp2s_v2.append(amp2_list)
    lss_v2.append(ls_list)
```

```python
amp2s_v5 = []
lss_v5 = []

for item in hypers_v5:
    amp2_list = [dict_['amp2'] for dict_ in item]
    ls_list = [dict_['ls'] for dict_ in item]
    amp2s_v5.append(amp2_list)
    lss_v5.append(ls_list)
```

```python
amp2s_v5_v2 = []
lss_v5_v2 = []

for item in hypers_v5_v2:
    amp2_list = [dict_['amp2'] for dict_ in item]
    ls_list = [dict_['ls'] for dict_ in item]
    amp2s_v5_v2.append(amp2_list)
    lss_v5_v2.append(ls_list)
```

### v1, v2, v5 ampllitude histograms

```python
for i in range(len(amp2s)):
    fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15,5))
    ax1.hist(amp2s[i])
    ax1.set_xlim(xmin=0, xmax=6)
    ax1.set_title("%f"%(np.std([item for item in amp2s[i]])))
    ax2.hist(amp2s_v2[i])
    ax2.set_xlim(xmin=0, xmax=6)
    ax2.set_title("%f"%(np.std([item for item in amp2s_v2[i]])))
    ax3.hist(amp2s_v5[i])
    ax3.set_xlim(xmin=0, xmax=6)
    ax3.set_title("%f"%(np.std([item for item in amp2s_v5[i]])))
    fig.suptitle("%d iter"%(i+5))
    plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot([np.std(amp2s[i]) for i in range(len(amp2s))],label='v1')
plt.plot([np.std(amp2s_v2[i]) for i in range(len(amp2s))],label='v2')
plt.plot([np.std(amp2s_v5[i]) for i in range(len(amp2s))],label='v5')
plt.legend()
plt.ylim(ymin=0, ymax=1)
plt.show()
```

### Plotting each iterations best acquisition fuction values variability

```python
np.argmax(np.average(acq_grids_v2[0],axis=0))
```

```python
acq_vals = []
acq_vals_v2 = []
acq_vals_v5 = []
acq_vals_v5_v2 = []
for iter_num in range(len(acq_grids_v2)):
    acq_vals.append([acq_grids[iter_num][mk][np.argmax(np.average(acq_grids[iter_num], axis=0))] for mk in range(len(acq_grids[iter_num]))])
    acq_vals_v2.append([acq_grids_v2[iter_num][mk][np.argmax(np.average(acq_grids_v2[iter_num], axis=0))] for mk in range(len(acq_grids_v2[iter_num]))])
    acq_vals_v5.append([acq_grids_v5[iter_num][mk][np.argmax(np.average(acq_grids_v5[iter_num], axis=0))] for mk in range(len(acq_grids_v5[iter_num]))])

for iter_num in range(len(acq_grids_v5_v2)):
    acq_vals_v5_v2.append([acq_grids_v5_v2[iter_num][mk][np.argmax(np.average(acq_grids_v5_v2[iter_num], axis=0))] for mk in range(len(acq_grids_v5_v2[iter_num]))])
```

```python
plt.figure(figsize=(10,5))
plt.plot([np.average(acq_vals[iter_num])for iter_num in range(len(acq_vals))], label='v1')
plt.plot([np.average(acq_vals_v2[iter_num])for iter_num in range(len(acq_vals_v2))], label='v2')
plt.plot([np.average(acq_vals_v5[iter_num])for iter_num in range(len(acq_vals_v5))], label='v5')
plt.plot([np.average(acq_vals_v5_v2[iter_num])for iter_num in range(len(acq_vals_v5_v2))], label='v5, amp v2')
plt.ylim(ymin=0, ymax=1)
plt.legend()
plt.show()
```
